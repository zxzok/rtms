### 指标变化与临床改善的相关性分析

```python
import pandas as pd

# 读取合并后的数据
df = pd.read_csv('/home/user/zhangyan/dominating_set/控制性指标与量表评分/frep_ac_score_diff.csv')

# 去掉 id 列，获取特征列和目标列
features = df.drop(columns=['id', 'HAMA_total_score_diff', 'HAMD17_total_score_diff', 'BPRS_total_score_diff'])
targets = df[['HAMA_total_score_diff', 'HAMD17_total_score_diff', 'BPRS_total_score_diff']]

# 计算相关性矩阵
corr_matrix = features.corrwith(targets['HAMA_total_score_diff']).to_frame(name='HAMA_corr')
corr_matrix['HAMD17_corr'] = features.corrwith(targets['HAMD17_total_score_diff'])
corr_matrix['BPRS_corr'] = features.corrwith(targets['BPRS_total_score_diff'])

# 添加一列表示方向（正相关/负相关）
corr_sign = corr_matrix.applymap(lambda x: '正相关' if x > 0 else '负相关')
corr_result = pd.concat([corr_matrix.round(3), corr_sign.add_suffix('_sign')], axis=1)

# 保存结果
corr_result.to_csv('/home/user/zhangyan/dominating_set/控制性指标与量表评分/score_correlation.csv')

# 设置相关性阈值
corr_threshold = 0.3
filtered_3 = corr_result[(corr_matrix.abs() >= corr_threshold).any(axis=1)]
filtered_3.to_csv('/home/user/zhangyan/dominating_set/控制性指标与量表评分/score_correlation_filtered(0.3).csv')

corr_threshold = 0.5
filtered_5 = corr_result[(corr_matrix.abs() >= corr_threshold).any(axis=1)]
filtered_5.to_csv('/home/user/zhangyan/dominating_set/控制性指标与量表评分/score_correlation_filtered(0.5).csv')
```

### Spearman 和 p 值

```python
from scipy.stats import spearmanr
import pandas as pd

# 读取合并后的数据
df = pd.read_csv('/home/user/zhangyan/dominating_set/控制性指标与量表评分/frep_ac_score_diff.csv')

# 去掉 id 列，获取特征列和目标列
features = df.drop(columns=['id', 'HAMA_total_score_diff', 'HAMD17_total_score_diff', 'BPRS_total_score_diff'])
targets = df[['HAMA_total_score_diff', 'HAMD17_total_score_diff', 'BPRS_total_score_diff']]

# Spearman 相关 + p 值
result_list = []
for col in features.columns:
    r1, p1 = spearmanr(features[col], targets['HAMA_total_score_diff'])
    r2, p2 = spearmanr(features[col], targets['HAMD17_total_score_diff'])
    r3, p3 = spearmanr(features[col], targets['BPRS_total_score_diff'])
    result_list.append([col, r1, p1, r2, p2, r3, p3])

# 转为 DataFrame
spearman_df = pd.DataFrame(result_list, columns=[
    'feature',
    'HAMA_spearman_corr', 'HAMA_p',
    'HAMD17_spearman_corr', 'HAMD17_p',
    'BPRS_spearman_corr', 'BPRS_p'
])

# 添加相关性方向
def sign(x):
    return '正相关' if x > 0 else '负相关'

spearman_df['HAMA_sign'] = spearman_df['HAMA_spearman_corr'].apply(sign)
spearman_df['HAMD17_sign'] = spearman_df['HAMD17_spearman_corr'].apply(sign)
spearman_df['BPRS_sign'] = spearman_df['BPRS_spearman_corr'].apply(sign)

# 保存结果
spearman_df.to_csv('/home/user/zhangyan/dominating_set/控制性指标与量表评分/score_spearman.csv', index=False)

# 也可以筛选出 |r| > 0.3 且 p < 0.05 的显著特征
filtered_spearman = spearman_df[
    ((spearman_df['HAMA_spearman_corr'].abs() >= 0.3) & (spearman_df['HAMA_p'] < 0.05)) |
    ((spearman_df['HAMD17_spearman_corr'].abs() >= 0.3) & (spearman_df['HAMD17_p'] < 0.05)) |
    ((spearman_df['BPRS_spearman_corr'].abs() >= 0.3) & (spearman_df['BPRS_p'] < 0.05))
]

filtered_spearman.to_csv('/home/user/zhangyan/dominating_set/控制性指标与量表评分/score_spearman_filtered(0.3_p0.05).csv', index=False)
```

### 画图

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from matplotlib.colors import LinearSegmentedColormap
from scipy.stats import gaussian_kde

# 创建保存目录
save_dir = '/home/user/zhangyan/dominating_set/控制性指标与量表评分/相关性/'
os.makedirs(save_dir, exist_ok=True)

# 为筛选后的显著特征绘制组合图
scatter_dir = f'{save_dir}combined_plots/'
os.makedirs(scatter_dir, exist_ok=True)

# 打印筛选出的显著特征数量
print(f"Found {len(filtered_spearman)} significant features (|r| > 0.3 and p < 0.05)")

# 定义颜色映射
cmap = LinearSegmentedColormap.from_list('custom_cmap', ['#1E88E5', '#FFC107', '#D81B60'])

for idx, row in filtered_spearman.iterrows():
    feature = row['feature']
    fig, axes = plt.subplots(3, 2, figsize=(18, 15), gridspec_kw={'width_ratios': [3, 1]})
    
    # 量表名称和颜色映射
    scales = ['HAMA', 'HAMD17', 'BPRS']
    colors = ['#1E88E5', '#FFC107', '#D81B60']
    
    for i, (scale, color) in enumerate(zip(scales, colors)):
        target_col = f'{scale}_total_score_diff'
        x = features[feature].values
        y = targets[target_col].values
        
        # 左侧：散点图 + 回归线 + KDE轮廓
        ax = axes[i, 0]
        
        # 1. 散点图
        ax.scatter(x, y, color=color, alpha=0.6, label=scale)
        
        # 2. 回归线
        m, b = np.polyfit(x, y, 1)
        ax.plot(np.sort(x), m*np.sort(x) + b, color=color, linestyle='-', linewidth=2)
        
        # 3. KDE轮廓
        try:
            nbins = 20
            k = gaussian_kde([x, y])
            xi, yi = np.mgrid[min(x):max(x):nbins*1j, min(y):max(y):nbins*1j]
            zi = k(np.vstack([xi.flatten(), yi.flatten()]))
            ax.contour(xi, yi, zi.reshape(xi.shape), 5, colors=color, alpha=0.5)
        except:
            # 如果KDE失败则跳过（可能是数据点太少）
            pass
        
        # 计算相关系数和p值
        r, p = spearmanr(x, y)
        
        # 设置图表标题和标签
        ax.set_title(f'{scale} vs {feature} (r={r:.2f}, p={p:.4f})', fontsize=12)
        ax.set_xlabel(feature, fontsize=10)
        ax.set_ylabel(f'{scale} score change', fontsize=10)
        ax.grid(True, alpha=0.3)
        ax.legend()
        
        # 右侧：箱线图 + 小提琴图
        ax_right = axes[i, 1]
        
        # 将数据分为两组：正向变化和负向变化
        positive_x = x[y > 0]
        negative_x = x[y <= 0]
        
        # 创建小提琴图和箱线图的组合
        violin_parts = ax_right.violinplot([positive_x, negative_x], 
                                           showmeans=False, showmedians=False, 
                                           showextrema=False)
        
        # 自定义小提琴图颜色
        for pc, c in zip(violin_parts['bodies'], [color, color]):
            pc.set_facecolor(c)
            pc.set_edgecolor('black')
            pc.set_alpha(0.3)
        
        # 添加箱线图
        box_parts = ax_right.boxplot([positive_x, negative_x], 
                                     positions=[1, 2], 
                                     patch_artist=True, 
                                     widths=0.3)
        
        # 自定义箱线图颜色
        for box, c in zip(box_parts['boxes'], [color, color]):
            box.set(facecolor=c, alpha=0.6)
            
        ax_right.set_xticklabels(['Improved', 'Worsened'])
        ax_right.set_title(f'{feature} distribution by\n{scale} outcome', fontsize=10)
        ax_right.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'{scatter_dir}combined_{idx+1}_{feature}.png', dpi=300)
    plt.close()

# 绘制综合p值分布图
plt.figure(figsize=(12, 8))

# 分别为每个量表绘制p值分布
scales = ['HAMA', 'HAMD17', 'BPRS']
colors = ['#1E88E5', '#FFC107', '#D81B60']

for scale, color in zip(scales, colors):
    p_col = f'{scale}_p'
    sns.kdeplot(spearman_df[p_col], label=f'{scale}', color=color, fill=True, alpha=0.3)
    sns.rugplot(spearman_df[p_col], color=color, alpha=0.5)

# 添加显著性阈值线
plt.axvline(0.05, color='red', linestyle='--', label='p=0.05 threshold')

# 添加累积分布
cum_data = []
cum_labels = []
for scale, color in zip(scales, colors):
    p_col = f'{scale}_p'
    sorted_p = np.sort(spearman_df[p_col].values)
    y = np.arange(1, len(sorted_p)+1) / len(sorted_p)
    cum_data.append((sorted_p, y, color, scale))
    cum_labels.append(f'{scale} (cumulative)')

# 添加次坐标轴绘制累积分布
ax2 = plt.gca().twinx()
for sorted_p, y, color, scale in cum_data:
    ax2.plot(sorted_p, y, color=color, linestyle=':', alpha=0.8, label=f'{scale} (cumulative)')
ax2.set_ylabel('Cumulative proportion')
ax2.grid(False)

# 设置图表属性
plt.title('P-value distributions for all features across scales')
plt.xlabel('p-value')
plt.ylabel('Density')
plt.xlim(0, 1)
plt.grid(True, alpha=0.3)

# 合并两个图例
lines1, labels1 = plt.gca().get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines1 + lines2, labels1 + labels2, loc='best')

plt.tight_layout()
plt.savefig(f'{save_dir}p_value_distribution_combined.png', dpi=300)
plt.close()
```
